```python
#importing required library
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#importing the dataset
myData=pd.read_csv('house_data.csv')
myData
```

```python
#checking the auto assigned values of dummy variables 
dummies=pd.get_dummies(myData.Location)
dummies
```

```python
#concatinating the dummy variable with dataset
merge=pd.concat([myData,dummies], axis='columns')
merge
```

```python
#removing chennai and location column from dataset
final_merge=merge.drop(['Location','Chennai'],axis='columns')
final_merge
```

```python
#Importing the linear regression model
from sklearn.linear_model import LinearRegression
createmodel=LinearRegression()
```

```python
#assign the values to x
x=final_merge.drop(['Price'],axis='columns')
x
```

```python
#assign the values to y
y=final_merge.Price
y
```

```python
#applying the linear regression model
createmodel.fit(x,y)
```

```python
#predicting the values 
createmodel.predict([[2000,2,3,2,2000,0,0,0,0,1,0]])
```

```python
#checking how perfect is our model
createmodel.score(x,y)
```
