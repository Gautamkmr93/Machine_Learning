```python
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the dataset
myData=pd.read_csv('house_data.csv')
myData
```

```python
#importing labelEncode and onehotencoder
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
#lableling the data set with the help of labelencoder
myDatalabel=myData
#myDatalabel
myDatalabel.Location=labelencoder.fit_transform(myDatalabel.Location)
myDatalabel
```

```python
#assining the x variable
x=myDatalabel[['Living_Space(sqft)','No_Of_Bedrooms','No_Of_Bathrooms','Condition(on scale of 5)',
               'Year_Build','Location']].values
x
```

```python
#assigning the y variable
y=myDatalabel.Price
y
```

```python
#importing onehotcoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(categorical_features=[5])
```

```python
x=ohe.fit_transform(x).toarray()
x
```

```python
x=x[:,1:]
x
```

```python
#importing the liner regression model
from sklearn.linear_model import LinearRegression
createModel=LinearRegression()
```

```python
#calculating the linear regression
createModel.fit(x,y)
```

```python
#predicting the price of house
createModel.predict([[0,0,0,0,1,0,2000,2,3,2,2000]])
```

```python
#checking how perfect is our model
createModel.score(x,y)
```
