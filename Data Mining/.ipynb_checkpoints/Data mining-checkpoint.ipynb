```python
data.fillna(method='ffill', inplace=True)
data.head()
```

```python
#fill NaN values using Zeros or using a fixed constant values
data['No_Of_Bedrooms']=data['No_Of_Bedrooms'].fillna(0)
data['No_Of_Bathrooms']=data['No_Of_Bathrooms'].fillna(100)
data['Living_Space(sqft)']=data['Living_Space(sqft)'].fillna(1000)
data['Condition(on scale of 5)']=data['Condition(on scale of 5)'].fillna(3)
data['Swimming Pool']=data['Swimming Pool'].fillna(data['Swimming Pool'].value_counts().index[0])
data['Location']=data['Location'].fillna(data['Location'].value_counts().index[0])
data['Price']=data['Price'].fillna(1000000)
data.head()
```

```python
data['Living_Space(sqft)']=data['Living_Space(sqft)'].fillna(data['Living_Space(sqft)'].mean())
data['No_Of_Bedrooms']=data['No_Of_Bedrooms'].fillna(data['No_Of_Bathrooms'].mean())
data['No_Of_Bathrooms']=data['No_Of_Bathrooms'].fillna(data['No_Of_Bathrooms'].mean())
data['Condition(on scale of 5)']=data['Condition(on scale of 5)'].fillna(data['Condition(on scale of 5)'].mean())
data['Swimming Pool']=data['Swimming Pool'].fillna(data['Swimming Pool'].value_counts().index[0])
data['Location']=data['Location'].fillna(data['Location'].value_counts().index[0])
data['Price']=data['Price'].fillna(data['Price'].mean())
data.head()
```

# Outliers

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
```

```python
data = pd.read_csv("data.csv",header =0)
data.head()
```

```python
sns.scatterplot(x='age',y='length_of_service',hue='is_promoted',data=data)
```

```python
# Box Plot to find more information about it's Quartiles and also detect Outliers
plt.figure(figsize = (5, 5))
plt.title('No_Of_Bedrooms')
sns.boxplot(y = data.No_Of_Bedrooms)
```

```python
#using violinplot
plt.figure(figsize = (5, 5))
plt.title('length_of_service')
sns.violinplot(y = data.length_of_service)
```

```python
Q1 = data.length_of_service.quantile(0.25)
Q3 = data.length_of_service.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
Data_outliers=data[(data.length_of_service<lower_limit)|(data.length_of_service>upper_limit)]
print(Data_outliers.shape)
data=data[(data.length_of_service>lower_limit)&(data.length_of_service<upper_limit)]#dropping the outliers from dataset
plt.figure(figsize = (5, 5))
plt.title('length_of_service')
sns.boxplot(y = data.length_of_service)
```

```python
#finding outliers using IQR formula 
#if "Datapoints<Q1-1.5 IQR where Q1=Quantile1" & "Datapoint>Q3+1.5 IQR where Q3=Quantile3"
Q1 = data.length_of_service.quantile(0.25)
Q3 = data.length_of_service.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
print(lower_limit,upper_limit)
mean_values=data['length_of_service'].mean()
data.loc[data.length_of_service > 8.0, 'length_of_service'] = np.nan
data['length_of_service']=data.length_of_service.fillna(mean_values)
plt.figure(figsize = (5, 5))
plt.title('length_of_service')
sns.boxplot(y = data.length_of_service)
```

# Dimension Reduction

```python
import numpy as np
import pandas as pd
data=pd.read_csv('New_Data.csv')
data.count()
```

```python
#Missing Values Ratio
print(data.isnull().sum()/len(data)*100)
print(data.isnull().sum())
```

```python
data=data.dropna()
print(data.isnull().sum())
```

```python
#Low Variance Filter
data.var()
```

```python
#High Correlation Filter
data.corr()
```

```python
#random forest
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1, max_depth=10)
data1=data.drop('Price',axis=1)
df=pd.get_dummies(data1)
model.fit(df,data.Price)
features = df.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-20:]  # top 20 features
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Score Values')
plt.show()
```
